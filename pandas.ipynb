MERGE

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
d2=pd.DataFrame({"C":[10,11,12,13],"B":[5,6,7,8]})
print(d1)
print()
print(d2)
```

```python
#common data will be merged
m1=pd.merge(d1,d2,on="B")
m2=pd.merge(d1,d2,how="inner")
m3=pd.merge(d1,d2,how="left")
print(m1)
print()
print(m2)
print()
print(m3)
```

```python
m2=pd.merge(d1,d2,how="inner",indicator=True)
m2
# indicator tells which value is present 
```

```python
dd1=pd.DataFrame({"A":[1,2,3,4],"B":[15,16,17,18]})
dd2=pd.DataFrame({"C":[10,11,12,13],"B":[5,6,7,8]})
mm2=pd.merge(dd1,dd2,right_index=True,left_index=True,suffixes=("name","id"))
mm2
```

CONCAT

```python
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([6,7,8,9],index=["a","b","c","d"])
s3=pd.concat([s1,s2])
s3
```

```python
ddr=pd.concat([dd1,dd2],axis=0)
ddc=pd.concat([dd1,dd2],axis=1)
print(ddr)
print()
print(ddc)
```

```python
#Outer join : UNION
#Inner join : INTERSECTION
ddi=pd.concat([dd1,dd2],axis=1,join='outer')
ddi
```

```python
ddc=pd.concat([dd1,dd2],axis=1,keys=["1st","2nd"])
ddc
```

JOIN

```python
j1=pd.DataFrame({"A":[1,2,3,4],"B":[4,5,6,7]})
j2=pd.DataFrame({"C":[11,12],"D":[24,25]})  #arrays in a particular dataframe shall be of same length
j3=j1.join(j2)
j3
```

```python
j4=j1.join(j2,how='inner')
j4
```

```python
j5=j1.join(j2,how='outer')
j5
```

```python
j6=j1.join(j2,how='left')
j6
```

```python
j7=j1.join(j2,how='right')
j7
```

```python
#if the name of colums are same,you can change them at runtime
j1=pd.DataFrame({"A":[1,2,3,4],"B":[4,5,6,7]})
j2=pd.DataFrame({"C":[11,12],"B":[24,25]})
j8=j1.join(j2,lsuffix="_11")
j8
```
